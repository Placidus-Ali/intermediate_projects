```python
import pandas as pd
```

```python
df = pd.read_csv("emission_factor_clean.csv")
df.head()
```

```python
def calculate_emissions(df):
    # Load the dataset
    df = pd.read_csv(df)
    
    # Extract country names (all columns except 'Activity')
    country_list = df.columns[1:].tolist()
    
    # Ask the user to choose a country
    print("Available countries:", ", ".join(country_list))
    country = input("Select a country: ").strip()
    
    if country not in country_list:
        print("Invalid country selection. Please select the right country.")
        return
    
    total_emission = 0
    
    # Iterate through each activity and ask for user input
    for index, row in df.iterrows():
        activity = row['Activity']
        factor = row[country]
        
        while True:
            try:
                user_input = float(input(f"Enter value for {activity}: "))
                break
            except ValueError:
                print("Invalid input. Please enter a numeric value.")
        
        total_emission += user_input * factor
    
    print(f"Total emission for {country}: {total_emission:.4f}")

# File path to dataset
df = "emission_factor_clean.csv"
calculate_emissions(df)
```
