Visualization Project

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri_data = sns.load_dataset('fmri')
```

```python
fmri_data.head()
```

Line Plot with Timepoint and Signal Column data

```python
sns.lineplot(x ='timepoint', y ='signal', data = fmri_data)
plt.show()
```

Mapping another column with a different color

```python
sns.lineplot(x ='timepoint', y ='signal', data = fmri_data, hue='event')
plt.show()
```

Adding some styles to the event column data

```python
sns.lineplot(x ='timepoint', y ='signal', data = fmri_data, hue='event', style='event')
plt.show()
```

Adding markers to value points

```python
sns.lineplot(x ='timepoint', y ='signal', data = fmri_data, hue='event', style='event', markers=True)
plt.show()
```

Visualization with BarPlot

```python
import pandas as pd
pokemon = pd.read_csv('C:/Users/STUTERN/Downloads/pokemon.csv')
```

```python
pokemon.head()
```

checking the speed of lengendary pokemon

```python
sns.barplot(x='is_legendary', y='speed', data=pokemon)
plt.show()
```

checking the weight of Pokemon

```python
sns.barplot(x='is_legendary', y='weight_kg', data=pokemon)
plt.show()
```

Let's add a hue to show pokemon generation

```python
sns.barplot(x='is_legendary', y='speed', data=pokemon, hue='generation')
plt.show()
```

Now let's use palette to assign different colors elements

```python
sns.barplot(x='is_legendary', y='speed', data=pokemon, palette='vlag')
plt.show()
```
