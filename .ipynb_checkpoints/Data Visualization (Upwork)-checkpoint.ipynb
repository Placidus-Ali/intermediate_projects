#### Objective

The goal is to compute various metrics and create charts based on this data.
To summarize the metrics, the following calculations are performed:

**Average Price**: Calculating the average price for each location.

**Average On-Peak Price**: Determining the average price during on-peak hours, which range from hour ending 7 to hour ending 22.

**Average Solar-Hours Price**: Finding the average price during solar hours, which span from hour ending 10 to hour ending 16.

**Price Percentiles (by year)**: Calculating price percentiles (P10, P25, P50, P75, P90) for each year to understand the distribution of prices.

**Average TB1, TB2, TB4**: Computing the average price differences between top and bottom hours of the day (TB1 represents the top 1 hour, TB2 represents the top 2 hours, and TB4 represents the top 4 hours).

**TB2 Arbitrage Percentiles (by year)**: Calculating TB2 arbitrage percentiles (P10, P25, P50, P75, P90) for each year, which represents the price difference between the top 2 hours and the bottom 2 hours of the day.

The charts generated based on the data and metrics include:


**Energy Price Averages**: A line chart displaying the average price, average on-peak price, and average solar-hours price over time. The chart allows toggling between monthly, quarterly, and yearly averages and enables selecting one or more locations.

**Price Percentiles**: Visual representation of price percentiles (P10, P25, P50, P75, P90) by year. The chart allows selecting one or more locations.

**Daily Arbitrage Averages**: A line chart showing the average TB1, TB2, and TB4 values over time. Similar to the previous charts, this one supports toggling between monthly, quarterly, and yearly averages and selecting one or more locations.

**TB2 Arbitrage Percentiles**: Visualization of TB2 arbitrage percentiles (P10, P25, P50, P75, P90) by year. It also allows selecting one or more locations.

**Daily Price Shape**: Shows the average price for each hour of the day for a selected location. The chart allows changing the location and toggling between specific months/quarters, as well as selectively displaying or hiding certain years.

**Average Price**

```python
# Calculating the average for each column
average_sp15 = df["SP15_DALMP"].mean()
average_np15 = df["NP15_DALMP"].mean()
average_ercot_n = df["ERCOT_N (DALMP)"].mean()

# Print the results
print("Average SP15 price:", average_sp15)
print("Average NP15 price:", average_np15)
print("Average ERCOT_N price:", average_ercot_n)

# Create a bar chart
labels = ['SP15', 'NP15', 'ERCOT_N']
average_prices = [average_sp15_price, average_np15_price, average_ercot_n_price]

plt.bar(labels, average_prices, color='blue')
plt.xlabel('Locations')
plt.ylabel('Average Price')
plt.title('Average Prices for SP15, NP15, and ERCOT_N')

# Put the price of each bar inside the bar
for i, price in enumerate(average_prices):
    plt.bar_label(plt.bar(labels[i], price), label=f"${price:.2f}", label_type='edge')
    
# Display the plot
plt.show()
```

**Average On Peak Price**

```python
# Extracting the hour component from the DATETIME column
df['Hour'] = pd.to_datetime(df['DATETIME']).dt.hour

# Filtering the rows where the hour is between 7 and 22 (inclusive)
on_peak_df = df[(df['Hour'] >= 7) & (df['Hour'] <= 22)]

# Computing the average on-peak price by taking the mean of the filtered price columns
average_on_peak_price = on_peak_df[['SP15_DALMP', 'NP15_DALMP', 'ERCOT_N (DALMP)']].mean()

# Plotting the average on-peak price graph
plt.figure(figsize=(8, 6))
bars = plt.bar(average_on_peak_price.index, average_on_peak_price, color='blue')
plt.title('Average On-Peak Price')
plt.xlabel('Locations')
plt.ylabel('Average Price')

# Put each price of the column on the bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f"${yval:.2f}", ha='center', va='bottom')

plt.show()
```

**Average Solar Hours Price**

```python
# Filter the rows where the hour is between 10 and 16 (inclusive)
solar_hours_df = df[(df['Hour'] >= 10) & (df['Hour'] <= 16)]

# Computting the average solar-hours price by taking the mean of the filtered price columns
average_solar_hours_price = solar_hours_df[['SP15_DALMP', 'NP15_DALMP', 'ERCOT_N (DALMP)']].mean()

# Plot the average solar-hours price graph
plt.figure(figsize=(8, 6))
bars = plt.bar(average_solar_hours_price.index, average_solar_hours_price, color='blue')
plt.title('Average Solar-Hours Price')
plt.xlabel('Locations')
plt.ylabel('Average Solar-Hours Price')

# Put each price of the column on the bar
for i, bar in enumerate(bars):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f"${average_solar_hours_price[i]:.2f}", ha='center', va='bottom')

plt.show()
```

**Price Percentiles (by year) (P10, P25, P50, P75, P90)**

```python
# Extractting the year component from the DATETIME column
df['Year'] = pd.to_datetime(df['DATETIME']).dt.year

# Grouping the data by year
grouped_data = df.groupby('Year')

# Computting the price percentiles (P10, P25, P50, P75, P90) for each year using the quantile function
price_percentiles = grouped_data[['SP15_DALMP', 'NP15_DALMP', 'ERCOT_N (DALMP)']].quantile([0.1, 0.25, 0.5, 0.75, 0.9])

# Displaying the price percentiles for each year
print("Price Percentiles (by year):")
print(price_percentiles)

# Plot the graph
plt.figure(figsize=(10, 6))
for column in price_percentiles.columns:
    plt.plot(price_percentiles.index.get_level_values('Year'), price_percentiles[column], label=column)
plt.title('Price Percentiles by Year')
plt.xlabel('Year')
plt.ylabel('Price')
plt.legend()
plt.show()
```

**Average TB1**

```python
# Extracting the date component from the DATETIME column
df['Date'] = pd.to_datetime(df['DATETIME']).dt.date

# Creating a new column that represents the difference between the top 1 hour and bottom 1 hour prices for each day
df['TB1'] = df.groupby('Date')['SP15_DALMP'].transform(lambda x: x.nlargest(1).values[0] - x.nsmallest(1).values[0])

# Computing the average TB1 by taking the mean of the new column
average_TB1 = df['TB1'].mean()

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['TB1'])
plt.title('Average TB1 Over Time')
plt.xlabel('Date')
plt.ylabel('Average TB1')
plt.xticks(rotation=45)
plt.show()
```

**Average TB2**

```python
# Creating a new column that represents the difference between the top 2 hours and bottom 2 hours prices for each day
df['TB2'] = df.groupby('Date')['SP15_DALMP'].transform(lambda x: x.nlargest(2).values.sum() - x.nsmallest(2).values.sum())

# Computing the average TB2 by taking the mean of the new column
average_TB2 = df['TB2'].mean()

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['TB2'])
plt.title('Average TB2 Over Time')
plt.xlabel('Date')
plt.ylabel('Average TB2')
plt.xticks(rotation=45)
plt.show()
```

**Average TB4**

```python
# Creating a new column that represents the difference between the top 4 hours and bottom 4 hours prices for each day
df['TB4'] = df.groupby('Date')['SP15_DALMP'].transform(lambda x: x.nlargest(4).values.sum() - x.nsmallest(4).values.sum())

# Computing the average TB4 by taking the mean of the new column
average_TB4 = df['TB4'].mean()

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['TB4'])
plt.title('Average TB4 Over Time')
plt.xlabel('Date')
plt.ylabel('Average TB4')
plt.xticks(rotation=45)
plt.show()
```

**TB2 Arbitrage Percentiles (by year)**

```python
# Computing the TB2 column representing the difference between the top 2 hours and bottom 2 hours prices for each day
df['TB2'] = df.groupby('Year')['SP15_DALMP'].transform(lambda x: x.nlargest(2).values.sum() - x.nsmallest(2).values.sum())

# Groupping the data by year and compute the TB2 arbitrage percentiles
tb2_percentiles = df.groupby('Year')['TB2'].quantile([0.1, 0.25, 0.5, 0.75, 0.9])

# Displaying the TB2 arbitrage percentiles for each year
print("TB2 Arbitrage Percentiles (by year):")
print(tb2_percentiles)
```

```python
# Computing the TB2 column representing the difference between the top 2 hours and bottom 2 hours prices for each day
df['TB2'] = df.groupby('Year')['SP15_DALMP'].transform(lambda x: x.nlargest(2).values.sum() - x.nsmallest(2).values.sum())

# Groupping the data by year and compute the TB2 arbitrage percentiles
tb2_percentiles = df.groupby('Year')['TB2'].quantile([0.1, 0.25, 0.5, 0.75, 0.9]).unstack()

# Plotting the graph
plt.figure(figsize=(10, 6))
for percentile in [0.1, 0.25, 0.5, 0.75, 0.9]:
    plt.plot(tb2_percentiles.index, tb2_percentiles[percentile], label=f'P{int(percentile*100)}')
plt.title('TB2 Arbitrage Percentiles by Year')
plt.xlabel('Year')
plt.ylabel('TB2 Arbitrage')
plt.legend()
plt.show()
```

**Energy Price Average**

```python
# Resampling the data based on the desired time period (monthly/quarterly/yearly) and compute the average prices
resampled_df = df.resample('M').mean()  # You can change 'M' to 'Q' for quarterly or 'Y' for yearly

# Selecting the locations you want to display on the chart
selected_locations = ['SP15_DALMP', 'NP15_DALMP', 'ERCOT_N (DALMP)']  # Update with the desired locations

# Plotting the line chart
plt.figure(figsize=(10, 6))

for location in selected_locations:
    plt.plot(resampled_df.index, resampled_df[location], label=location)

plt.title('Energy Price Averages')
plt.xlabel('Time')
plt.ylabel('Average Price ($/MWh)')
plt.legend()
plt.show()
```

**Price Percentiles**

```python
# Groupping the data by year
grouped_df = df.groupby('Year').quantile([0.1, 0.25, 0.5, 0.75, 0.9])

# Selecting the locations you want to display on the chart
selected_locations = ['SP15_DALMP', 'NP15_DALMP', 'ERCOT_N (DALMP)']  # Update with the desired locations

# Plotting the line chart
plt.figure(figsize=(10, 6))

for location in selected_locations:
    x = grouped_df.index.get_level_values('Year')
    y = grouped_df.loc[:, location]

    plt.plot(x, y, label=location)

plt.title('Price Percentiles')
plt.xlabel('Year')
plt.ylabel('Price Percentiles ($/MWh)')
plt.legend()
plt.show();
```

**Daily Arbitrage Averages**

```python
# Computing Metrics
tb1 = df.resample('D').apply(lambda x: x.nlargest(1).mean() - x.nsmallest(1).mean())
tb2 = df.resample('D').apply(lambda x: x.nlargest(2).mean() - x.nsmallest(2).mean())
tb4 = df.resample('D').apply(lambda x: x.nlargest(4).mean() - x.nsmallest(4).mean())

#Creating Charts
# Line chart showing metrics 5-7 (TB1, TB2, TB4) over time
plt.plot(tb1.index.date, tb1.values, label='TB1')
plt.plot(tb2.index.date, tb2.values, label='TB2')
plt.plot(tb4.index.date, tb4.values, label='TB4')
plt.xlabel('Date')
plt.ylabel('Price ($/MWh)')
plt.title('Daily Arbitrage Averages')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

# Line chart with monthly averages
monthly_tb1 = tb1.resample('M').mean()
monthly_tb2 = tb2.resample('M').mean()
monthly_tb4 = tb4.resample('M').mean()

plt.plot(monthly_tb1.index, monthly_tb1.values, label='Monthly TB1')
plt.plot(monthly_tb2.index, monthly_tb2.values, label='Monthly TB2')
plt.plot(monthly_tb4.index, monthly_tb4.values, label='Monthly TB4')
plt.xlabel('Date')
plt.ylabel('Price ($/MWh)')
plt.title('Monthly Arbitrage Averages')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

**TB2 Arbitrage Percentiles**

```python
# Computting TB2 Arbitrage Percentiles
df['DATETIME'] = pd.to_datetime(df['DATETIME'])  # Convert DATETIME column to datetime data type
df['Year'] = df['DATETIME'].dt.year  # Extract the year from the datetime column

# Calculating TB2 (top 2 hours - bottom 2 hours) for each day
df['TB2'] = df.groupby(df['DATETIME'].dt.date)['SP15_DALMP'].transform(lambda x: x.nlargest(2).mean() - x.nsmallest(2).mean())

# Groupping by year and calculate percentiles
percentiles = df.groupby('Year')['TB2'].quantile([0.1, 0.25, 0.5, 0.75, 0.9]).unstack()

# Creating the chart
plt.figure(figsize=(10, 6))
sns.lineplot(data=percentiles, markers=True)
plt.title('TB2 Arbitrage Percentiles by Year')
plt.xlabel('Year')
plt.ylabel('TB2 Arbitrage Value')
plt.legend(['P10', 'P25', 'P50', 'P75', 'P90'])

# Show the chart
plt.show()
```

**Daily Price Shape ($/MWh)**

```python
# Plotting average price for each hour of the day, showing each year as a different line
location = 'SP15_DALMP'
hourly_prices = df[location].groupby([df.index.year, df.index.hour]).mean().unstack()

years = hourly_prices.index
for year in years:
    plt.plot(hourly_prices.columns, hourly_prices.loc[year], label=str(year))

plt.xlabel('Hour of the Day')
plt.ylabel('Price ($/MWh)')
plt.title('Daily Price Shape')
plt.legend()
plt.show()
```
